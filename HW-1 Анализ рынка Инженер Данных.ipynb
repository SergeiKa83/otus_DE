```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
from matplotlib import cm
import random
import matplotlib.colors as mcolors

data = pd.read_excel("hw-1_data.xlsx", header=1)
data.sample(5)
```

# Рынок DE: Количество вакансий, отрасли компаний, размер компаний

```python
df = data.groupby("Регион")["Компания"].count()
labels = df.index.tolist() 
sizes = df.values.tolist()

def func(pct, allvals):
    absolute = int(pct/100.*np.sum(allvals))
    return "{:.1f}%\n({:d} вакансий)".format(pct, absolute)

fig1, ax1 = plt.subplots(figsize=(7,7), subplot_kw=dict(aspect="equal"))
 
wedges, texts, autotexts = ax1.pie(sizes, labels=labels, autopct=lambda pct: func(pct, sizes), textprops=dict(color="k", fontsize=14))
ax1.axis('equal')
ax1.legend(loc='upper left', bbox_to_anchor=(1.0, 1.0))
ax1.set_title("Распределение вакансий по регионам в данных", fontdict={'fontsize': 20})
plt.show()
```

```python
df = data.groupby("Отрасль компании")["Компания"].count()
labels = df.index.tolist() 
sizes = df.values.tolist()

def func(pct, allvals):
    absolute = int(pct/100.*np.sum(allvals))
    return "{:d}".format(absolute)

fig1, ax1 = plt.subplots(figsize=(7,7), subplot_kw=dict(aspect="equal"))
 
wedges, texts, autotexts = ax1.pie(sizes, labels=labels, autopct=lambda pct: func(pct, sizes), textprops=dict(color="k", fontsize=14))
ax1.axis('equal')
ax1.legend(loc='upper left', bbox_to_anchor=(1.7, 1.0))
ax1.set_title("Распределение вакансий по отраслям", fontdict={'fontsize': 20}, pad = 40.)
plt.show()
```

```python
df = data.groupby("Размер компании")["Компания"].count()
labels = df.index.tolist() 
sizes = df.values.tolist()

def func(pct, allvals):
    absolute = int(pct/100.*np.sum(allvals))
    return "{:.1f}%\n({:d} вакансий)".format(pct, absolute)

fig1, ax1 = plt.subplots(figsize=(7,7), subplot_kw=dict(aspect="equal"))
 
wedges, texts, autotexts = ax1.pie(sizes, labels=labels, autopct=lambda pct: func(pct, sizes), textprops=dict(color="k", fontsize=14))
ax1.axis('equal')
ax1.legend(loc='upper left', bbox_to_anchor=(1.0, 1.0))
ax1.set_title("Распределение вакансий по размеру компании", fontdict={'fontsize': 20}, pad = 40.)
plt.show()
```
